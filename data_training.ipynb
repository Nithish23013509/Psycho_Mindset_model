```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import HumanMessage, SystemMessage
from datasets import load_dataset
import random

# Load the DialogSum dataset
dataset = load_dataset("knkarthick/dialogsum")

# Function to get a random dialogue
def get_random_dialogue():
    # Get a random example from the dataset
    random_example = random.choice(dataset['train'])
    return random_example['dialogue']  # Return the dialogue

# Initialize the chatbot
ollama_base_url = "http://test1.dgx.saveetha.in:8080/v1"
chat = ChatOpenAI(base_url=ollama_base_url, openai_api_key="your_api_key", model="llama3.1")

# Define a function to interact with the Iron Man chatbot
def chat_with_iron_man(user_input):
    # Initialize messages
    messages = []

    # Set the system message
    system_message = "You are Iron Man, also known as Tony Stark. You're witty, confident, and a bit sarcastic. Answer the user's questions as Iron Man would."
    messages.append(SystemMessage(content=system_message))

    # Add user input as a human message
    messages.append(HumanMessage(content=user_input))

    # Get a random dialogue snippet for context
    iron_man_response = get_random_dialogue()
    messages.append(HumanMessage(content=iron_man_response))  # Adding dialogue context

    # Invoke the chatbot
    response = chat.invoke(messages)

    # Parse the response
    parser = StrOutputParser()
    bot_response = parser.invoke(response)

    return bot_response

# Example interaction with the Iron Man chatbot
if __name__ == "__main__":
    print("Iron Man Bot is ready! Type 'exit' to end the chat.")
    while True:
        user_input = input("You: ")
        if user_input.lower() in ['exit', 'quit']:
            print("Iron Man: Goodbye, genius! Remember, I have an arsenal of suits. Don't make me send them after you!")
            break
        response = chat_with_iron_man(user_input)
        print(f"Iron Man: {response}")
```
